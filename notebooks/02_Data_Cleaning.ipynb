```python
# ============================================================
# Notebook 02 : Data Cleaning
# Healthcare Readmission Analytics
# Objectives
# ------------------------------------------------------------
# 1. Identify missing values
# 2. Replace invalid values
# 3. Remove unnecessary columns
# 4. Fix data types
# 5. Save cleaned dataset
# ============================================================

import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)
```

```python
# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv("../data/raw/diabetic_data.csv")
```

```python
# ============================================================
# Dataset Overview Before Cleaning
# ============================================================

print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")

df.head()
```

```python
# ============================================================
# Replace Invalid Missing Values
# ============================================================

df.replace("?", np.nan, inplace=True)
```

```python
# ============================================================
# Missing Value Analysis
# ============================================================

missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": (df.isnull().sum() / len(df) * 100).round(2)
})

missing = missing[missing["Missing Values"] > 0]

missing.sort_values(by="Percentage", ascending=False)
```

```python
# ============================================================
# Drop Columns with Excessive Missing Values
# ============================================================

columns_to_drop = [
    "weight",
    "max_glu_serum",
    "A1Cresult",
    "medical_specialty",
    "payer_code"
]

df.drop(columns=columns_to_drop, inplace=True)

print("Columns dropped successfully.")
print(f"Remaining Columns : {df.shape[1]}")
```

```python
# ============================================================
# Check Remaining Missing Values
# ============================================================

remaining_missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": (df.isnull().sum()/len(df)*100).round(2)
})

remaining_missing = remaining_missing[
    remaining_missing["Missing Values"] > 0
]

remaining_missing
```

```python
# ============================================================
# Fill Remaining Missing Values
# ============================================================

columns_to_fill = ["race", "diag_1", "diag_2", "diag_3"]

for column in columns_to_fill:
    df[column] = df[column].fillna(df[column].mode()[0])

print("Remaining missing values filled successfully.")
```

```python
# ============================================================
# Verify Missing Values
# ============================================================

missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": (df.isnull().sum()/len(df)*100).round(2)
})

missing = missing[missing["Missing Values"] > 0]

missing
```

```python
# ============================================================
# Check Duplicate Records
# ============================================================

duplicates = df.duplicated().sum()

print(f"Duplicate Rows : {duplicates}")
```

```python
# ============================================================
# Dataset Shape After Cleaning
# ============================================================

print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")
```

```python
# ============================================================
# Save Clean Dataset
# ============================================================

df.to_csv("../data/processed/diabetic_data_clean.csv", index=False)

print("Clean dataset saved successfully.")
```

```python
# ============================================================
# Data Cleaning Summary
# ============================================================

print("=" * 50)
print("DATA CLEANING SUMMARY")
print("=" * 50)

print(f"Original Rows      : 101766")
print(f"Remaining Rows     : {df.shape[0]}")

print()

print("Dropped Columns")
print("----------------")

for column in columns_to_drop:
    print(f"• {column}")

print()

print("Filled Missing Values")
print("----------------------")

for column in columns_to_fill:
    print(f"• {column}")

print()

print(f"Final Columns : {df.shape[1]}")

print()
print("Notebook 02 Completed Successfully")
```
